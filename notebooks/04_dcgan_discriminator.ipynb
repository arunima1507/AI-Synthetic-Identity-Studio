```python
import torch
import torch.nn as nn
```

```python
class Discriminator(nn.Module):
    
    def __init__(self):

        super().__init__()

        self.main = nn.Sequential(

            nn.Conv2d(
                3, 64, 4, 2, 1,
                bias=False
            ),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(
                64, 128, 4, 2, 1,
                bias=False
            ),
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(
                128, 256, 4, 2, 1,
                bias=False
            ),
            nn.BatchNorm2d(256),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(
                256, 512, 4, 2, 1,
                bias=False
            ),
            nn.BatchNorm2d(512),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(
                512, 1, 4, 1, 0,
                bias=False
            ),

            nn.Sigmoid()
        )

    def forward(self, x):
        return self.main(x)
```

```python
discriminator = Discriminator()

print(discriminator)
```

```python
fake_image = torch.randn(1,3,64,64)

print(fake_image.shape)
```

```python
output = discriminator(fake_image)

print(output.shape)
print(output)
```
